```python
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv("nfl_pbp.csv")
```

```python
df = df[['game_id','posteam','game_date','rush_attempt','pass_attempt','qtr','down']]
df = df.dropna()
```

```python
plays_down = df[(df.down <= 3) & (df.qtr < 5) & (df.posteam == 'NE')]
ax = plays_down.groupby(['game_date'])['pass_attempt','rush_attempt'].sum().plot.line(color=['red','blue'],figsize=(20,9),rot=0,fontsize=16)
ax.set_title("Play Calling By Season - New England Patriots", fontsize=24)

ax.set_xlabel("Season", fontsize=20)
ax.set_xticklabels(item.get_text()[:4] for item in ax.get_xticklabels())
ax.set_ylabel("# Plays", fontsize=20)
ax.set_alpha(1.0)
```
